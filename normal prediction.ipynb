```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('train.csv')
```

```python
data.describe()
```

```python
y=data.pop('Survived')
y.head()
```

```python
y
```

```python
numeric=list(data.dtypes[data.dtypes != 'object'].index)
data[numeric].head()
```

```python
data['Age'].fillna(data.Age.mean(),inplace=True)
```

```python
data.tail()
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(data[numeric],y)
```

```python
print('train Accuracy : : :',accuracy_score(y,model.predict(data[numeric])))
```

```python
test=pd.read_csv('test.csv')
```

```python
test[numeric].head()
```

```python
test['Age'].fillna(data.Age.mean(),inplace=True)
test=test[numeric].fillna(test.mean()).copy()
```

```python
y_predict=model.predict(test[numeric])
```

```python
y_predict
```

```python
submissions=pd.DataFrame({
    "PassengerId" : test["PassengerId"],
    "Survived" : y_predict
})
submissions.to_csv('titanicId.csv',index=False)
```
